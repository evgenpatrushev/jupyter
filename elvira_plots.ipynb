```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 999)

%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
df.head()
```

## - кто сказал нет и их паспортичка (последние данные: пол, возраст, город, доход) 

```python
filter = df['Чи знаєте Ви про інсування здорових снеків/батончиків?']=='Ні'
data1 = df[filter]
print('Количество записей = ', len(data1))
data1 = data1[['Ваша стать',
       'Ваш вік ', 'У якому населеному пункті Ви проживаєте?',
       'Ваш місячний рівень доходу']].sort_values(by=['Ваша стать'])
data1
```

```python
labels = {x for x in set(data1['Ваша стать'].values) if x==x}
d = []
for label in labels:
    d.append(len(data1[data1['Ваша стать'] == label]))
explode = (0.05, 0.05, 0.05) 

fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(d, labels=labels, explode=explode, autopct='%2.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
labels = {x for x in set(data1['Ваш вік '].values) if x==x}
d = []
for label in labels:
    if label != 'nan':
        d.append(len(data1[data1['Ваш вік '] == label]))
explode = (0.05, 0.05, 0.05, 0.05, 0.05) 

fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(d, labels=labels, explode=explode, autopct='%2.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
