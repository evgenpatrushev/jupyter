$f(x)=x*cos2x,  a=0,  b=1$

```python
def f(x):
    return x * np.cos(2 * x)
```

```python
x = np.arange(-1, 15, 0.01)
y = f(x)
plt.plot(x, y)
plt.plot([-1, 15], [0,0],'black')
plt.plot([0,0],[15,-15],'black')
print('График фукнции f(x)')
```

```python
def check(a, b, N):
    
    if type(a) is not int or type(a) is not float:
        print('type a must int or float')
        return 1    
    
    if type(b) is not int or type(b) is not float:
        print('type b must int or float')
        return 1
    
    if type(N) is not int:
        print('type N must int')
        return 1
    
    if -1000 > a > 1000 or -1000 > b > 1000:
        print('a/b must be between interval [-1000, 1000]')
        return 1
    
    if a >= b:
        print('a must be lower than b')
        return 1
    
    if b - a >= 100:
        print('difference between b and a must be lower than 100')
        return 1
    
    if 0 > N >= 1000:
        print('N must be lower 1000')
        return 1
    
    return 0
    
```

```python
def integral(N=1, a=0, b=1):
    
    if check(a=a, b=b, N=N):
        return 
    
    h = round((b-a) / N, 2)
    
    x = np.arange(a, b+h, h)
    if x[-1] > 1:
        x[-1] = 1
    elif x[-1] < 1:
        np.append(x, 1)
    y = f(x)

    x_p = np.arange(a, b, 0.01)
    y_p = f(x_p)
    
    plt.plot(x_p, y_p, 'b')
    plt.plot(x, y, 'g')
    for x_i, y_i in zip(x, y):
        plt.plot([x_i, x_i], [0, y_i], 'g-')
    plt.plot([a, b], [0,0],'black')
    plt.plot([0,0],[max(y_p),min(y_p)],'black')

    return sum([(y[i] + y[i+1]) * (y[i+1] - y[i]) / 2  for i in range(len(x)-1)])
```

```python
a, b, N = 0, 1, 1
S = integral(N, a, b)
```
