$f(x)=x*cos2x,  a=0,  b=1$

```python
def f(x):
    return x * np.cos(2 * x)

def diff_f(a, b, n):
    x_val = np.linspace(a, b, 1000)
    x = Symbol('x')
    y = x*cos(2*x)
    yprime = y.diff(x).diff(x)
    arr = [abs(y.diff(x).subs(x, xi)) for xi in x_val]
    m = max(arr)
    return (b-a)**3 * m / (12 * n**2)
```

```python
x = np.arange(-1, 15, 0.01)
y = f(x)
plt.plot(x, y)
plt.plot([-1, 15], [0,0],'black')
plt.plot([0,0],[15,-15],'black')
print('Графік фукнції f(x)')
```

```python
def check(a, b, N):
    
    if type(a) is not int and type(a) is not float:
        print('type a must int or float')
        return 1    
    
    if type(b) is not int and type(b) is not float:
        print('type b must int or float')
        return 1
    
    if type(N) is not int:
        print('type N must int')
        return 1
    
    if -1000 > a  or a > 1000 or -1000 > b or b > 1000:
        print('a/b must be between interval [-1000, 1000]')
        return 1
    
    if a >= b:
        print('a must be lower than b')
        return 1
    
    if b - a >= 100:
        print('difference between b and a must be lower than 100')
        return 1
    
    if 0 > N or N >= 1000:
        print('N must be lower 1000')
        return 1
    
    return 0
    
```

```python
def integral(N=1, a=0, b=1):
    
    if check(a=a, b=b, N=N):
        return None, None
    
    h = round((b-a) / N, 100)
    
    x = np.arange(a, b+h, h)
    if x[-1] > 1:
        x[-1] = 1
    elif x[-1] < 1:
        np.append(x, 1)
    y = f(x)

    x_p = np.arange(a, b, 0.01)
    y_p = f(x_p)
    
    plt.plot(x_p, y_p, 'b')
    plt.plot(x, y, 'g')
    for x_i, y_i in zip(x, y):
        plt.plot([x_i, x_i], [0, y_i], 'g-')
    plt.plot([a, b], [0,0],'black')
    plt.plot([0,0],[max(y_p),min(y_p)],'black')
    
    error = diff_f(a,b,N)
    print('Похибка по модулю менше або дорівнює = ', error)
    
    return_val = sum([(y[i] + y[i+1]) * (x[i+1] - x[i]) / 2  for i in range(len(x)-1)]) 
    return return_val, error+return_val
```

```python
a, b, N = 0, 1, 1
S, error_S = integral(N, a, b)
print('значення інтегралу S = ', S)
print('S + error = ', error_S)
```
